# Kernel PCA

- Non-linear dimensionality reduction through use of kernels

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.datasets import make_circles
from sklearn.decomposition import PCA, KernelPCA
```

```python
np.random.seed(0)
X, y = make_circles(n_samples=400, factor=0.3, noise=0.05)
```

```python
kpca = KernelPCA(kernel='rbf', fit_inverse_transform=True, gamma=10)
x_kpca = kpca.fit_transform(X)
x_back = kpca.inverse_transform(x_kpca)

pca = PCA()
x_pca = pca.fit_transform(X)
```

## Normal Plot

```python
plt.figure()
plt.title('Original Space')
reds = y == 0
blues = y == 1

plt.scatter(X[reds, 0], X[reds, 1], c='red', s=20, edgecolor='k')
plt.scatter(X[blues, 0], X[blues, 1], c='blue', s=20, edgecolor='k')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$');
```

```python
plt.figure()
plt.title('Projection By PCA')
reds = y == 0
blues = y == 1

plt.scatter(x_pca[reds, 0], x_pca[reds, 1], c='red', s=20, edgecolor='k')
plt.scatter(x_pca[blues, 0], x_pca[blues, 1], c='blue', s=20, edgecolor='k')
plt.xlabel('First Component')
plt.ylabel('Second Component');
```

```python
plt.figure()
plt.title('Projection By KPCA')
reds = y == 0
blues = y == 1

plt.scatter(x_kpca[reds, 0], x_kpca[reds, 1], c='red', s=20)
plt.scatter(x_kpca[blues, 0], x_kpca[blues, 1], c='blue', s=20)
plt.xlabel('First Component in space induced by $\phi$')
plt.ylabel('Second Component');
```
