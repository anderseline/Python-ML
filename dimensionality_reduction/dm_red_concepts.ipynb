# Dimensionality Reduciton

## Principal Component Analysis (PCA)
- Linear dimensionality reduction using singular value decomposition of the data to project it to a lower dimensionsal space
- Uses orthogonal transformation (think linear algebra)
- Converts correlated features (predictors) into linearly uncorrelated features called principal components
- Number of principal components <= number of features
- First principal component exmplains the largest possible variance
- Collection of components are called vectors
- Each subsequent component has the highest variance subject to the restriction that it must be orthogonal to the preceding components
- Sensitive to scaling

Method
1. Eigenalue decomposition of data covariance (or correlation) matrix (essentially removes dependent or correlated data through use of eigenvalues)
2. Singular value decomposition of a data matrix (after mean centering) the data matrix for each attribute

Output
1. Component scores, called factor scores (transformed variable values)
2. Loadings (weights)

## Linear Discriminant Analysis (LDA)
- Commonly used as a dimensionality reduction technique in preprocessing for pattern-classification 
- Projects dataset onto a lower-dimensional space with good class-separability in order to avoid overfitting and reduce computational cost
- Locate boundaries around clusters of classes
- Projects datapoints on a line


```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
sns.set()
```

```python
rnd_num = np.random.RandomState(42)
X = np.dot(rnd_num.rand(2,2), rnd_num.randn(2, 500)).T
```

```python
X
```

```python
X [:,0] = -X[:,0]
```

```python
plt.scatter(X[:,0], X[:,1])
plt.axis('equal');
```

## PCA Identification

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(X)
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_)
```

```python
print(pca.explained_variance_ratio_)
```

```python
plt.scatter(X[:,0], X[:,1], alpha=0.3)

for k, v in zip(pca.explained_variance_, pca.components_):
    vec = v * 3 * np.sqrt(k)
    ax = plt.gca()
    arrowprops = dict(
        arrowstyle='<-', linewidth=1, shrinkA=0, shrinkB=0
    )
    ax.annotate('',pca.mean_, pca.mean_ + vec, arrowprops=arrowprops)
    ax.text(-0.90, 1.2, 'PC1', ha='center', va='center', rotation=-42, size=12)
    ax.text(-0.1, -0.6, 'PC2', ha='center', va='center', rotation=50, size=12)
plt.axis('equal');
```

## Note orthogonal nature of PC comoponents
## Length of each denotes significance

```python
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
```

```python
X.shape
```

```python
X_pca.shape
```

```python
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:,0], X[:,1], alpha=0.2)
plt.scatter(X_new[:,0], X_new[:,1], alpha=0.8)
plt.axis('equal');
# orange line is the PCA (reduced to a single dimension)
```
