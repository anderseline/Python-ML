# Linear Regression

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.datasets import load_boston
import seaborn as sns   
sns.set(color_codes=True)
import matplotlib.pyplot as plt 
import warnings
boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['MEDV'] = boston_dataset.target
```

```python
from sklearn.linear_model import LinearRegression
```

```python
boston.head()
```

```python
# use number of rooms independent variable
# reshape between -1 and 1 for sklearn usability 
X = boston['RM'].values.reshape(-1, 1)
```

```python
y = boston['MEDV'].values
```

```python
# instantiate model as a linear regression from sklearn import 
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
# slope
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(10,8))
sns.regplot(X, y)
plt.xlabel('Average Rooms per Home')
plt.ylabel('Median Home Value in $1000s')
plt.show()
```

```python
sns.jointplot(x='RM', y='MEDV', data=boston, kind='reg', height=10)
plt.show()
```

```python
# can inspect model predictions for home value at each value for RM
rm_value = 5
model.predict(np.array([rm_value]).reshape(1,-1))
```
